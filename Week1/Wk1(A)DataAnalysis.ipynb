Dealing with missing values - Filling: Đối với phương án thế giá trị mới vào các ô trống, ta
có thể sử dụng các giá trị mean, median... của cột dữ liệu tương ứng để thay thế. Ví dụ, có
một vài hàng có Revenue mang giá trị null, ta có thể gán cho nó giá trị trung bình

```python
# Dealing with missing values - Filling
revenue_mean = data_indexed['Revenue (Millions)'].mean() # old value, before replacing NaN values
print ("The mean revenue is: ", revenue_mean )

# We can fill the null values with this mean revenue
data_indexed['Revenue (Millions)'].fillna(revenue_mean, inplace = True )
new_revenue_mean = data_indexed ['Revenue (Millions)'].mean()
print ("The NEW mean revenue is: ", new_revenue_mean )
```

apply() functions: Apply functions được dùng khi ta muốn thực thi một hàm nào đó lên các
hàng trong bảng dữ liệu

```python
# Classify movies based on ratings
def rating_group(rating):
  if rating >= 7.5:
      return 'Good '
  elif rating >= 6.0:
    return 'Average '
  else:
    return 'Bad '

# creating a new variable in the dataset to hold the rating cat
data['Rating_category'] = data ['Rating'].apply(rating_group) # new collumn "Rating_category"
data[['Title','Director','Rating','Rating_category']].head(5)
```
